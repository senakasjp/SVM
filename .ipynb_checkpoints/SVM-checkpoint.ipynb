```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn import svm

import matplotlib.pyplot as plt
import seaborn as sns;sns.set(font_scale=1.2)
recipes=pd.read_csv('iris.csv')
recipes.head()

sns.lmplot('sepal_width','petal_width',data=recipes,hue='species',palette='Set1',fit_reg=False,scatter_kws={'s':70})

sepal_width_petal_width=recipes[['sepal_width','petal_width']].as_matrix()
type_label=np.where(recipes['species']=='setosa',0,1)
type_label2=np.where(recipes['species']=='versicolor',0,1)
type_label3=np.where(recipes['species']=='virginica',0,1)
type_label0=[]
for i in range(len(type_label)):
    if type_label[i]==0:
        type_label0.append(0)
    elif type_label2[i]==0:
        type_label0.append(1)
    elif type_label3[i]==0:
        type_label0.append(2)
# print(type_label0)
model=svm.SVC(kernel='linear')
model.fit(sepal_width_petal_width,type_label0)
# w=model.coef_[0]
# a=-w[0]/w[1]
# xx=np.linspace(5,30)
# yy=a*xx-(model.intercept_[0])/w[1]

# b=model.support_vectors_[0]
# yy_down=a*xx+(b[1]-a*b[0])
# b=model.support_vectors_[-1]
# yy_up=a*xx+(b[1]-a*b[0])

def predict_flower(sepal_width,petal_width):
    if (model.predict([[sepal_width,petal_width]]))==0:
        print ("setosa")
    elif (model.predict([[sepal_width,petal_width]]))==1:
        print ("versicolor")
    else:
        print ("virginica")
predict_flower(3,2)
```
